### üí° DICAS GERAIS PARA RESOLVER:

**Sobre LISTAS:**

- lista.append(x) - adiciona no final

- lista.remove(x) - remove um item

- len(lista) - conta quantos itens

- max(lista) / min(lista) - maior/menor valor

**Sobre LOOPS:**

- for item in lista: - para cada item

- while condicao: - repete enquanto for verdade

- break - sai do loop

**Sobre CONDICIONAIS:**

- if condicao: - executa se for verdade

- elif outra_condicao: - sen√£o, se outra for verdade

- else: - se nenhuma for verdade

**Sobre TIPOS:**

- int() - converte para n√∫mero inteiro

- float() - converte para n√∫mero decimal

- str() - converte para texto

### üí° DICAS PARA DICION√ÅRIOS:
**Acessando valores:**

- dicionario["chave"] - acessa um valor

- dicionario.get("chave") - acessa com seguran√ßa

- dicionario.keys() - todas as chaves

- dicionario.values() - todos os valores

- dicionario.items() - pares chave-valor

**Modificando:**

- dicionario["nova_chave"] = valor - adiciona/atualiza

- dicionario.update(outro_dict) - atualiza m√∫ltiplos valores

- del dicionario["chave"] - remove
